<h1 align="center">Image Classification using Neural Network</h1>

```python
import tensorflow as ft
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from tensorflow import keras
print(ft.__version__)
```

```python
(x_train,y_train) , (x_test,y_test)=keras.datasets.mnist.load_data()
```

```python
x_train=x_train/255
```

```python
x_test=x_test/255
```

```python
len(x_train)
```

```python
len(x_test)
```

```python
x_train[0].reshape
```

```python
plt.matshow(x_train[0])
```

<h1 align="center">TO convert into Nurans it must be in Single Diamention</h1>

```python
x_train_fleatend=x_train.reshape(len(x_train),28*28)
```

```python
x_test_fleatend=x_test.reshape(len(x_test),28*28)
```

```python
model=keras.Sequential([
keras.layers.Dense(10,input_shape=(784,),activation='sigmoid')
])
model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)
model.fit(x_train_fleatend,y_train,epochs=5)
```

```python
model.evaluate(x_test_fleatend,y_test)
```

```python
y_predicted=model.predict(x_test_fleatend)
```

```python
y_predicted[0]
```

```python
y_predicted=model.predict(x_test_fleatend)
y_predicited_labels=[np.argmax(i) for i in y_predicted]
```

```python
y_predicited_labels[:5]
```

```python
np.argmax(y_predicted[1])
```

<h1 align="center">Confusion Matrix to test Accuracy </h1>

```python
cm=ft.math.confusion_matrix(labels=y_test, predictions=y_predicited_labels)
```

```python
cm
```

```python
plt.figure(figsize=(10,7))
sns.heatmap(cm,annot=True,fmt='d')
plt.xlabel('predicted')
plt.ylabel('True')
```

```python
model=keras.Sequential([
keras.layers.Dense(100,input_shape=(784,),activation='relu'),
    keras.layers.Dense(10,activation='sigmoid')
])
model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)
model.fit(x_train_fleatend,y_train,epochs=5)
```

```python
model.evaluate(x_test_fleatend,y_test)
```

<h1 align="center"> Digits Classifications using Nural Network to maximize Accuracy</h1>

```python
import tensorflow as tf
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from tensorflow import keras 
```

# Load Dataset of digits

```python
(x_train,y_train),(x_test,y_test)=keras.datasets.mnist.load_data()
```

```python
x_train_fleatend=x_train.reshape(len(x_train),28*28)
x_test_fleatend=x_test.reshape(len(x_test),28*28)
```

```python
print(len(x_test))
print(len(x_train))
```

```python
# Make Nural Network Dense Layers to perform Classification
```

```python
model= keras.Sequential([
    keras.layers.Dense(10,input_shape=(28*28,),activation='sigmoid'),
])

model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)
model.fit(x_train_fleatend,y_train,epochs=5)
```

```python
model.evaluate(x_test_fleatend,y_test)
```

```python
model.evaluate(x_test_fleatend,y_test)
```

```python
y_train[0:5]
```

<h1 align="center">Confusion Matrix to Test Accuracy Error</h1>

```python
y_predicted=model.predict(x_test_fleatend)
y_predicited_labels=[np.argmax(i) for i in y_predicted]
cm=ft.math.confusion_matrix(labels=y_test, predictions=y_predicited_labels)
plt.figure(figsize=(10,7))
sns.heatmap(cm,annot=True,fmt='d')
plt.xlabel('predicted')
plt.ylabel('True')
```
